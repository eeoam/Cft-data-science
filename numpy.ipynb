# Introduction

```python
import numpy as np
```

# Arrays

```python
a = np.array([1,2,3])
print(a)
```

```python
b = np.array([[9.0, 8.0, 7.0], [6.0, 5.0, 4.0]])
print(b)
```

```python
# Get dimension
print(a.ndim)
print(b.ndim)
```

```python
# Get shape
print(a.shape)
print(b.shape)
```

```python
# Get type
print([a.dtype, b.dtype])
```

```python
# Get size in bytes
print([a.itemsize , b.itemsize])
```
